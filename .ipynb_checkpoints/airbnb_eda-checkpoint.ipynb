```python
# Load in the dataframe for airbnb.
listings = pd.read_csv('listings.csv', delimiter=',')
```

#### Data description and data cleaning
For this exploratory data analysis we are only interested in private rooms or in entire home/apparment rooms times, so the data is only going to display information about those two room types. <br>
For doing all the analysis on the airbnb data we are going to be using the following columns:

neighbourhood_group_cleansed <- neighbourhood where the listing is located <br>
latitude <- latitude location of the property <br>
longitude <- longitude location of the property <br>
acccommodates <- number of guests the listing can accommodate <br>
price -> listing price in US dolars per night <br>
number_of_reviews <- number of reviews for 2023 <br>
distance_from_city_centre <- a new created column using geopy.distance that represents the distance in km between the airbnb location and Plasa Catalunya (considered as the city centre of Barcelona for this analysis)

The first step is looking through the data, making sure all the variables we want to use are correct (no typing errors) and that there is no outliers

```python
listings = listings[(listings.room_type == 'Entire home/apt')]
listings.sort_values(by=['id'])

#estimate numbers of days booked
bookings = pd.read_csv('calendar.csv', delimiter=',')
bookings['available'] = bookings['available'].map({"t": 1, "f": 0})
days_booked = 365- bookings.groupby(["listing_id"]).available.sum()
days_booked_df =days_booked.reset_index()
days_booked_df.sort_values(by=['listing_id'])
# Remove the $ from the price variables.
listings['price'] = listings['price'].replace('[$,]', '', regex=True).astype(float)
listings['days_booked'] = days_booked_df['available']
#listings.head()
#listings.info()
#listings.isna().sum()

listings["income per year"] = listings['days_booked'] * listings['price'] 
```

```python
# some data cleaning first, we are going to look at the distribution of prices and see whether 
# there are dodgey values
sns.histplot(data=listings, x="price", bins=30).set(title='Price distribution for entire home/apparments')
# clearly we can see that values above 2100 seems clear outliers and outisde of interest of this, as they don't 
# represent the bulk of the data
listings.groupby("accommodates")["price"].describe().round(2)
```

By looking at the above data distribution, we can see:
- There is some clear outliers on prices for maximum values, where listing prices sit way above the 75% percentile, very clear outliers (e.g. listings with a price of 84999 , 4500, 3528) and also we can conclude that values above 2000 are very unlikely and they are not representative for answering our question. 
- There is as well minimum price values that seem incorrect, for example an entire home/apparment that can accomodate 5 to 8 people being listed as $10 per night. 

In order to capture the most representative data for the goal of this analysis we are going to consider outliers any data point outside the range of 1.5 times the IQR below the 25th percentile or above the 75th percentile



```python
# calculate IQR for column price
Q1 = listings['price'].quantile(0.25)
Q3 = listings['price'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = listings[(listings['price'] < Q1 - threshold * IQR) | (listings['price'] > Q3 + threshold * IQR)]

#create new dataframe with outliers removed

listings_cleaned = listings.drop(outliers.index)

sns.histplot(data = listings_cleaned, x = "price" , bins=30).set(title='Entire home price distribution')
```

The plot above show the price distribution for entire home listings and private room listings, the distribution seems right right skewed, so when reporting price statistics it will be better to use the median rather than the mean. 

```python
# TODO: calculate distance in m to city centre of barcelona
import geopy.distance
#Geopy is a Python library that simplifies the calculation of geographic distances between two points.
# Geopy uses the Vincenty distance uses more accurate ellipsoidal models, such as WGS-84, and is implemented in geopy. For example,
city_center_coordinates = (41.3879555, 2.1700572 )

#create a new column that is distance to city centre, city centre is considered as Plasa Catalunya longitude and latitude location
listings_cleaned['Distance_to_city_centre'] = listings_cleaned.apply(lambda x: geopy.distance.distance((x.latitude, x. longitude), city_center_coordinates).km, axis=1)
```

#### What are the hot areas in Barcelona for airbnb rentals displayed visually?


```python
from folium.plugins import HeatMap
map = folium.Map(location = [listings.latitude[30],listings.longitude[30]], 
                 tiles = "CartoDB positron",   zoom_start=15)

df_heatmap = listings[['latitude','longitude', 'price']]
#we can't have NaN on a heatmap

df_heatmap = df_heatmap.dropna()

df_heatmap
map.add_child(
    HeatMap(
        df_heatmap,
        min_opacity=0.2,
        radius = 8,
        blur = 6
    )
)
```
